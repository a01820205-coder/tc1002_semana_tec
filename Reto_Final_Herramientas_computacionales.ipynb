```python
X = df[['PURCHASES', 'CASH_ADVANCE']].copy()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)
print(labels)
print(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X["PURCHASES"], X["CASH_ADVANCE"],
            c=labels, cmap="pink", alpha=0.6, edgecolor="pink")
plt.xlabel("Compras")
plt.ylabel("Anticipo en efectivo")
plt.title("Clusters de Clientes")
plt.show()
```

Se pueden observar dos grupos, pero que tiene dos derivados.

 1.Clientes que piden poco anticipo, pero que casi no realizan compras (1.1) y otros de formas moderada(1,2)

2.-Clientes que piden alto anticipo, pero hay unos que tienen muchas compras (2.1) y otros que casi no realizan compras.(2,2)

Por lo tanto sería bueno mantener a los clientes tipo 1.2 y 2.1, ya que son clientes que generan crédito, generan ingresos y que organizan sus gastos. Sin embargo, los clientes tipo 1.1 y 2.2, son aquellos que no son organizados con sus gastos, pueden caer en deudas y no generan ingresos, o directamente son clientes inactivos.
Por lo tanto a los clientes que se buscan mantener, se les ofrecería una tarjeta que al tener un anticipo y gasto similares y que puedan mantenerlo, una tarjeta que en base a su anticipo se les de el 5% de descuento en futuras compras.
A los que pueden ser conflictivos, se les puede ofrecer una tarjeta que tenga un limite de anticipo y que este vaya aumentado según sus compras, esto con el objetivo de organizar sus gatos y evitar riesgos de deuda.
